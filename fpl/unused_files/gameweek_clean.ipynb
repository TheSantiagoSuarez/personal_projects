```python
import pandas as pd
```

```python
#creating all the names for all the csv we want to import, one per gameweek
filenames = [5]
```

Gameweek Squads

```python
#make them into dataframes
dfs = [pd.read_csv(f"raw_csv\{str(filename)}.csv") for filename in filenames]
```

```python
#rename some columns and drop useless columns and blank rows
for i in dfs:
    i.rename(columns={'Unnamed: 0':'Fielded','Unnamed: 1':'Player','Unnamed: 20':'Manager'}, inplace=True )
    i.rename(columns={'GS':'Goals','A':'Assists','S':'Saves','B':'Bonus'}, inplace=True )

    i.drop(columns=["BPS","I","C","T","II"], inplace=True)
    i.dropna(how='all',inplace=True)
```

```python
#solve the "player" column
for i in dfs:
    split = i["Player"].str.split("\n",expand=True)

    i["Name"] = split.iloc[:, 0]
    i["Team"] = split.iloc[:, 1]
    i["Team"] = i["Team"].str.slice(stop=-3)
    i.drop(columns="Player", inplace=True)
```
