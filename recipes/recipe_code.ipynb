```python
import urllib.request
import http.cookiejar

url = "https://www.modernhoney.com/the-best-snickerdoodle-cookie-recipe/"

# Set headers
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
    'Accept-Encoding': 'none',
    'Accept-Language': 'en-US,en;q=0.8',
    'Connection': 'keep-alive'
}

# Create cookie jar
cj = http.cookiejar.CookieJar()
opener = urllib.request.build_opener(urllib.request.HTTPCookieProcessor(cj))
urllib.request.install_opener(opener)

# Create request
req = urllib.request.Request(url, headers=headers)

try:
    with urllib.request.urlopen(req) as response:
        content = response.read().decode('utf-8')
        print(content)
except urllib.error.HTTPError as e:
    print(f"HTTP Error: {e.code}")
    print(e.read().decode('utf-8'))
```

```python
print(content)
```

```python
with open("recipe.html", "w", encoding="utf-8") as f:
    f.write(content)
```

```python
from requests_html import HTMLSession

url = "https://www.modernhoney.com/the-best-snickerdoodle-cookie-recipe/"
session = HTMLSession()
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
}
try:
    response = session.get(url, headers=headers)
    response.html.render(timeout=20)  # Executes JS
    print(response.html.html)
except Exception as e:
    print(f"Error: {e}")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.modernhoney.com/the-best-snickerdoodle-cookie-recipe/"
url = "https://bromabakery.com/brown-butter-brownies/"

# Set headers to mimic a real browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',
    'Accept-Language': 'en-US,en;q=0.9'
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Try to extract recipe title
    title = soup.find('h1', class_='wprm-recipe-name')
    print("Recipe Title:", title.text.strip() if title else "Not found")

    # Extract ingredients
    print("\nIngredients:")
    ingredients = soup.find_all(class_='wprm-recipe-ingredient')
    for ing in ingredients:
        print("-", ing.get_text(strip=True))

    # Extract instructions
    print("\nInstructions:")
    instructions = soup.find_all(class_='wprm-recipe-instruction-text')
    for i, step in enumerate(instructions, 1):
        print(f"{i}. {step.get_text(strip=True)}")
else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
with open("recipe.html", "w", encoding="utf-8") as f:
    f.write(soup.text)
```
